```python
import numpy as np

# Constants
timeStep = 100           # years
waterDepth = 4000        # meters
L = 1350                 # Watts/m2
albedo = 0.3
epsilon = 1
sigma = 5.67E-8          # W/m2 K^4

# Calculate the incoming heat flux
heatIn = L * (1 - albedo) / 4

# Define heat capacity of the surface (J/m2 K)
heatCapacity = waterDepth * 4.2E6  # J/m2 K

# Input number of time steps
nSteps = int(input())

# Initial values
time_list = [0]
temperature_list = [0]  # Initial temperature in Kelvin
heatContent = temperature_list[0] * heatCapacity

# Simulation loop
for step in range(nSteps):
    # Calculate the outgoing heat flux
    heatOut = epsilon * sigma * pow(temperature_list[-1], 4)
    
    # Calculate the change in heat content
    dHeatContent_dt = heatIn - heatOut
    heatContent += dHeatContent_dt * timeStep * 3.154e+7  # Convert years to seconds
    
    # Update temperature
    temperature = heatContent / heatCapacity
    temperature_list.append(temperature)
    time_list.append((step + 1) * timeStep)
    
# Final calculations
final_heatOut = epsilon * sigma * pow(temperature_list[-1], 4)

# Print the final temperature and heat flux
print(f"{temperature_list[-1]} {final_heatOut}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
timeStep = 100          # years
waterDepth = 4000       # meters
L = 1350                # Watts/m2
albedo = 0.3
epsilon = 1
sigma = 5.67E-8         # W/m2 K^4
nSteps = int(input())   # Number of time steps

# Calculating heat capacity
heatCapacity = waterDepth * 4.2E6  # J/m2 K, assuming the heat capacity of water

# Initial values
time_list = [0]
temperature_list = [0]  # Starting with 0 K
heatContent = temperature_list[0] * heatCapacity

# Simulation loop
for step in range(nSteps):
    # Calculate the incoming and outgoing heat fluxes
    heatIn = L * (1 - albedo) / 4
    heatOut = epsilon * sigma * pow(temperature_list[-1], 4)
    
    # Update heat content
    dHeatContent_dt = heatIn - heatOut
    heatContent += dHeatContent_dt * timeStep * 3.154e+7  # Convert years to seconds
    
    # Update temperature
    temperature = heatContent / heatCapacity
    temperature_list.append(temperature)
    time_list.append((step + 1) * timeStep)

# Calculate final outgoing heat flux for last temperature
heatOut = epsilon * sigma * pow(temperature_list[-1], 4)

# Output the final temperature and heat flux
print(temperature_list[-1], heatOut)

# Plotting temperature vs time
plt.plot(time_list, temperature_list)
plt.xlabel('Time (years)')
plt.ylabel('Temperature (K)')
plt.title('Temperature of a Naked Planet Over Time')
plt.grid(True)
plt.show()
```
